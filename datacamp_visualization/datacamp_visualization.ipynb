```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
df=pd.read_csv('./percent-bachelors-degrees-women-usa.csv', header=0)
```

```python
df.head()
```

```python
physical_sciences = df.loc[:,'Physical Sciences'].values
```

```python
computer_science = df.loc[:,'Computer Science'].values
```

```python
year = df.loc[:,'Year'].values
```

```python
plt.plot(year, physical_sciences, color='blue')
plt.plot(year, computer_science, color='red')
```

```python
# create plot axes for first line plot
plt.axes([0.05,0.05,0.425,0.9])

plt.plot(year, physical_sciences, color='blue')

# create plot axes for  second plot
plt.axes([0.525,0.05,0.425,0.9])

plt.plot(year, computer_science, color='red')

plt.show()
```

```python
# create a figure with 1x2 subplot and make the left subplot active
plt.subplot(1,2,1)
# this means 1 row two columns, 1st one
# create plot axes for first line plot

plt.plot(year, physical_sciences, color='blue')

plt.subplot(1,2,2)

plt.plot(year, computer_science, color='red')

plt.tight_layout()
# used tight layout for better spacing.

plt.show()
```

```python
# create a subplot with 2x2 subplot layout and make the top left subplot active
# two rows, two columns, number of the plot
plt.subplot(2,2,1)
plt.plot(year, physical_sciences, color='blue')

plt.subplot(2,2,2)
plt.plot(year, computer_science, color='red')

education = health = df['Education'].values

plt.subplot(2,2,3)
plt.plot(year, education, color='green')

health = df['Health Professions'].values

plt.subplot(2,2,4)
plt.plot(year, health, color='yellow')

plt.tight_layout()
plt.show()
```

```python
plt.plot(year, computer_science, color='red')
plt.plot(year, physical_sciences, color='blue')
plt.xlabel('Year')
plt.ylabel('Degrees awarded to women(%)')
plt.xlim([1990,2010])
plt.ylim([0,50])

plt.title('Degrees awarded to women (1990-2010)\nComputer Science(red)\nPhysical Sciences(blue)')

plt.show()
plt.savefig('xlim_and_ylim.png')
```

```python
plt.plot(year, computer_science, color='red', label='Computer Science')
plt.plot(year, physical_sciences, color='blue', label='Physical Sciences')

plt.legend(loc="lower center")

plt.xlabel('year')
plt.ylabel('Enrollment(%)')
plt.title("Undergraduate enrollment of women")
plt.show()
```

```python
plt.plot(year, computer_science, color='red', label='Computer Science')
plt.plot(year, physical_sciences, color='blue', label='Physical Sciences')

plt.legend(loc="lower right")

cs_max = computer_science.max()

yr_max = year[computer_science.argmax()]

plt.annotate('Maximum', xy=(yr_max, cs_max), arrowprops=dict(facecolor='black'))

plt.xlabel('Year')
plt.ylabel('Enrollment (%)')
plt.title('Undergraduate enrollment of women')
plt.show()
```

```python
import numpy as np
```

```python
u=np.linspace(-2, 2, 41)
v=np.linspace(-1, 1, 21)
```

```python
X, Y = np.meshgrid(u, v)
```

```python
X.shape
```

```python
Y.shape
```

```python
Z=np.sin(3*np.sqrt(X**2+Y**2))
```

```python
plt.pcolor(Z)
plt.show()
plt.savefig('sine_mesh.png')
```

```python
plt.subplot(2,2,1)
plt.contour(X, Y, Z)

plt.subplot(2,2,2)
plt.contour(X, Y, Z, 20)

plt.subplot(2,2,3)
plt.contourf(X,Y,Z)

plt.subplot(2,2,4)
plt.contourf(X, Y, Z, 20)

plt.tight_layout()
plt.show()
```
