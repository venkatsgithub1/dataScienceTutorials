```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Visualizing non linear regression

```python
# x has data from -5 to +5 with step 0.1
x = np.arange(-5.0, 5.0, 0.1)
x
```

### y=mx+c
### y = 2x + 3

```python
y = 2*x+3
y
```

### create noise using normal distribution

## Cubic function graph

```python
x = np.arange(-5.0,5.0,0.1)
y=1*(x**3)+1*(x**2)+1*(x)+3
y_noise=20*np.random.normal(size=x.size)
y_data=y+y_noise
```

```python
plt.plot(x,y_data, 'bo')
plt.plot(x,y,'r')
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

### Quadratic

```python
x = np.arange(-5.0, 5.0, 0.1)

y=np.power(x,2)
y_noise=np.random.normal(size=x.size)
ydata=y+y_noise
plt.plot(x,ydata,'bo')
plt.plot(x,y,'r')
plt.ylabel('Dependent Variables')
plt.xlabel('Independent Variables')
plt.show()
```

### Exponential

```python
x=np.arange(-5,5,0.1)
y=np.exp(x)
plt.plot(x,y)
plt.ylabel('Dependent Variables')
plt.xlabel('Independent Variables')
plt.show()
```

### Logarithmic

```python
X=np.arange(-5,5,0.1)
Y=np.log(X)

plt.plot(X,Y)
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

### Sigmoidal / Logistic

```python
X=np.arange(-5,5,0.1)

Y=1-4/(1+np.power(3,X-2))

plt.plot(X,Y)
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

# Non Linear Regression Example
## Data set: China GDP from 1960 to 2014

```python
import numpy as np
import pandas as pd

df = pd.read_csv('china_gdp.csv')
df.head(10)
```

## Plotting the dataset

```python
plt.figure(figsize=(8,5))
xdata,ydata=(df.loc[:,"Year"],df.loc[:,"Value"])
plt.plot(xdata,ydata,'ro')
plt.ylabel('GDP')
plt.xlabel('Year')
plt.show()
```

## Choosing a model

```python
X = np.arange(-5.0, 5.0, 0.1)
y=1.0/(1.0+np.exp(-X))

plt.plot(X,y)
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

```python
def sigmoid(x, beta1, beta2):
    y=1/(1+np.exp(-beta1*(x-beta2)))
    return y
```

```python
beta1 = 0.1
beta2=1990

Y_pred = sigmoid(xdata, beta1, beta2)

plt.plot(xdata,Y_pred*150000000000000)
plt.plot(xdata,ydata,'ro')
```

## Normalizing data

```python
xdata=xdata/max(xdata)
ydata=ydata/max(ydata)
```

```python
from scipy.optimize import curve_fit
popt, pcov = curve_fit(sigmoid, xdata, ydata)
print("beta1={}, beta2={}".format(popt[0],popt[1]))
```

```python
x=np.linspace(1960,2015,55)
x=x/max(x)
plt.figure(figsize=(8,5))
y=sigmoid(x,*popt)
plt.plot(xdata,ydata,'ro',label='data')
plt.plot(x,y,linewidth=3.0,label='fit')
plt.legend(loc='best')
plt.ylabel('GDP')
plt.xlabel('Year')
plt.show()
```

## Split data into train/test

```python
msk = np.random.rand(len(df))<0.8
train_x=xdata[msk]
train_y=ydata[msk]

test_x=xdata[~msk]
test_y=ydata[~msk]

# building model using training set
popt,pcov = curve_fit(sigmoid, train_x, train_y)

# prediction using testing
y_hat = sigmoid(test_x, *popt)

y_hat
```

## Evaluation

```python
print("mean absolute error", np.mean(np.absolute(y_hat-test_y)))
print("Residual Sum of Squares", np.mean((y_hat-test_y)**2))

from sklearn.metrics import r2_score
print("r2 score:", r2_score(y_hat, test_y))
```
