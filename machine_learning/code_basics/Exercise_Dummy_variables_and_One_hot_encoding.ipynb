```python
import pandas as pd
```

```python
df = pd.read_csv('carprices_label_encoding.csv')
```

```python
df
```

```python
dummies = pd.get_dummies(df['Car Model'])
```

```python
merged = pd.concat([df, dummies], axis=1)
```

```python
merged
```

```python
merged = merged.drop('Car Model', axis=1)
```

```python
merged
```

```python
merged.drop('Mercedez Benz C class', axis=1, inplace=True)
```

```python
X=merged.drop('Sell Price($)', axis=1)
X
```

```python
y=merged.loc[:,['Sell Price($)']]
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
# benz + 45000 mile-age, 4 yr
model.predict([[45000,4,0,0]])
```

```python
# bmw + 86000 mile-age, 7 yr
model.predict([[86000,7,0,1]])
```

```python
model.score(X, y)
```
