# Section 2

## Find missing values

```python
co2_levels=pd.read_csv('data/ch2_co2_levels.csv')
print(co2_levels.head())
```

```python
# converting datestamp column to datetime64
co2_levels['datestamp']=pd.to_datetime(co2_levels['datestamp'])
print(co2_levels.dtypes)
co2_levels=co2_levels.set_index('datestamp')
```

```python
co2_levels.head()
```

```python
# checking type
co2_levels.dtypes
```

```python
# missing values
co2_levels.isnull().sum()
```

## Handle missing values

```python
# filling missing values using fillna
co2_levels=co2_levels.fillna(method='bfill')
```

```python
#find out how many are still missing
print(co2_levels.isnull().sum())
```

## Display rolling averages

```python
# 52 weeks rolling avg
ma=co2_levels.rolling(window=52).mean()
```

```python
ma.head(52)
```

```python
mstd=co2_levels.rolling(window=52).std()
```

```python
mstd[mstd.index.year==1958]
```

### Using empirical rule to calculate lower and upper bounds

```python
ma['upper']=ma['co2']+(2*mstd['co2'])
```

```python
ma['lower']=ma['co2']-(2*mstd['co2'])
```

```python
# store ax object
ax=ma.plot(linewidth=0.8, fontsize=6)
# adding x label
ax.set_xlabel('Date', fontsize=10)
# adding y label
ax.set_ylabel('CO2 levels in Mauai Hawaii from 1950 to 2001', fontsize=10)
# set title
ax.set_title('Rolling mean and variance of CO2 levels in Mauai Hawaii from 1958 to 2001', fontsize=10)
plt.show()
```

## Display aggregated values

```python
# get month for dates in index
index_month=co2_levels.index.month
index_month
```

```python
# mean co2_levels by month
mean_co2_levels_by_month=co2_levels.groupby(index_month).mean()
# plot mean values
mean_co2_levels_by_month.plot(fontsize=6)
# legend with font size
plt.legend(fontsize=10)
# show graph
plt.show()
```

## Compute numerical summaries

```python
# summary stats of co2_levels dataframe
co2_levels.describe()
```

```python
co2_levels['co2'].min()
```

```python
co2_levels['co2'].max()
```

## Boxplots and Histograms

```python
# generating box plot
ax=co2_levels.boxplot()
# set x label
ax.set_xlabel('CO2', fontsize=10)
# set y label
ax.set_ylabel('Boxplot CO2 levels in Maui Hawaii', fontsize=10)
# add legend
plt.legend(fontsize=10)
plt.show()
```

```python
# histogram
ax=co2_levels.plot(kind='hist',bins=50,fontsize=6)
# setting x label
ax.set_xlabel('CO2', fontsize=10)
# setting y label
ax.set_ylabel('Histogram of CO2 levels in Maui Hawaii', fontsize=10)
# legend
plt.legend(fontsize=10)
plt.show()
```

## Density plots

```python
# plotting kde plot
ax=co2_levels.plot(kind='density', linewidth=4, fontsize=6)
# xlabel
ax.set_xlabel('CO2', fontsize=10)
# y label
ax.set_ylabel('Density plot of CO2 levels in Maui Hawaii', fontsize=10)
# showing plot
plt.show()
```

# Section 3

## Autocorrelation in time series data

```python
# importing statsmodels.graphics
from statsmodels.graphics import tsaplots
# fig object returned by tsaplots plot_acf
fig=tsaplots.plot_acf(co2_levels['co2'], lags=24)
plt.show()
```

## Partial autocorrelation in time series data

```python
# store returned fig
fig=tsaplots.plot_pacf(co2_levels['co2'], lags=24)
plt.show()
```

## Time series decomposition

```python
# importing statsmodels api
import statsmodels.api as sm
# perform timeseries decomposition
decomposition = sm.tsa.seasonal_decompose(co2_levels)
# figure out all functions of decomposition
print(dir(decomposition))
```

### From above we can see, we can use seasonal, trend, resid for error, plot is also part

```python
decomposition.seasonal
```

## Individual components plots

### Trend

```python
# getting trend data from decomposition
trend = decomposition.trend
print(trend)
```

```python
# getting ax from plot
ax = trend.plot(figsize=(12,6), fontsize=6)
# x label
ax.set_xlabel('Date',fontsize=10)
# title
ax.set_title('Seasonal component the CO2 time-series', fontsize=10)
# show
plt.show()
```

## Add summary statistics to time series plot

```python
# store ax
ax=meat.plot(fontsize=9,linewidth=1,figsize=(16,10))
# x lbl
ax.set_xlabel('Date', fontsize=6)
# get meat mean values
meat_mean=meat.describe().loc[['mean']]
# add summary statistics to plot
ax.table(cellText=meat_mean.values, colWidths=[0.15]*len(meat_mean.columns),rowLabels=meat_mean.index,colLabels=meat.columns,loc='top')
# legend
ax.legend(loc='upper center', bbox_to_anchor=(0.5,0.95),ncol=3,fontsize=9)
# show plot
plt.show()
```

## Plotting multiple time series plots on subplots

```python
# plots on 2 rows and 4 columns
meat.plot(subplots=True, layout=(2,4), sharex=False,sharey=False,colormap='viridis',fontsize=7,legend=False,linewidth=0.2,figsize=(16,9))
# show plots
plt.show()
```

```python
# better view
# plots on 2 rows and 4 columns
meat.plot(subplots=True, layout=(4,2), sharex=False,sharey=False,colormap='viridis',fontsize=6,legend=False,linewidth=0.3,figsize=(16,9))
# show plots
plt.show()
```

## Computing correlation between time series

```python
# spearman corr between beef and pork
print(meat[['beef','pork']].corr(method='spearman'))
print(meat[['beef','pork']].corr(method='spearman').loc['beef','pork'])
```

```python
# using pearson coefficient on pork, veal, turkey
print(meat[['pork','veal','turkey']].corr(method='pearson'))
corr_mat=meat[['pork','veal','turkey']].corr(method='pearson')
# between veal and pork
print(corr_mat.loc['pork','veal'])
# between pork and turkey
print(corr_mat.loc['pork','turkey'])
# between veal and turkey
print(corr_mat.loc['turkey','veal'])
```

## Visualizing correlation matrices

```python
# import seaborn
import seaborn as sns
# correlation matrix of meat
corr_meat=meat.corr(method='spearman')
# heatmap using sns
sns.heatmap(corr_meat,annot=True,linewidths=0.4,annot_kws={"size":10})
# xticks
plt.xticks(rotation=90)
#y ticks
plt.yticks(rotation=0)
# show plot
plt.show()
```

## Clustered heatmaps

```python
# cluster map and store obj
fig=sns.clustermap(corr_meat, row_cluster=True, col_cluster=True, figsize=(10,10))
# x labels setup
plt.setp(fig.ax_heatmap.xaxis.get_majorticklabels(),rotation=90)
# y labels setup
plt.setp(fig.ax_heatmap.yaxis.get_majorticklabels(),rotation=0)
# show
plt.show()
```

# Section 5

## Explore jobs dataset

```python
# reading jobs csv
jobs=pd.read_csv('data/ch5_employment.csv')
# check the data
print(jobs.head(5))
# check datatypes
print(jobs.dtypes)
```

```python
# convert datestamp column to datetime64 ns
jobs['datestamp']=pd.to_datetime(jobs['datestamp'])
# set datestamp column as index
jobs=jobs.set_index('datestamp')
```

```python
# check data
jobs.head()
```

```python
# check nulls
print(jobs.isnull().sum())
```

## Using boxplots to describe timeseries

```python
# boxplot
jobs.boxplot(fontsize=6,vert=False)
# show
plt.show()
# describe
print(jobs.describe())
# timeseries with highest mean
print(jobs.describe().loc['mean'].max())
# highest std
print(jobs.describe().loc['std'].max())
```

## Plotting all timeseries columns

```python
# get subset of data
jobs_subset=jobs[['Finance','Information','Manufacturing','Construction']]
# check data
print(jobs_subset.head())
# graphs on 2x2
jobs_subset.plot(subplots=True,layout=(2,2),sharex=False,sharey=False,linewidth=0.7,fontsize=6,legend=False,figsize=(16,9))
# show plot
plt.show()
```

## Annotating significant events

```python
# plotting all columns of timeseries
# store ax obj
ax=jobs.plot(colormap='Spectral',fontsize=6,linewidth=0.8)
# x label
ax.set_xlabel('Date')
# y label
ax.set_ylabel('Unemployment Rate',fontsize=10)
# title
ax.set_title('Unemployment rate of U.S. workers by industry',fontsize=10)
# legend location
ax.legend(loc='center left', bbox_to_anchor=(1.0,0.5))

#annotations
# axvline 1
ax.axvline('2001-07-01',color='blue',linestyle='--',linewidth=0.8)
# axvline 2
ax.axvline('2008-09-01',color='blue',linestyle='--',linewidth=0.8)

# plt
plt.show()
```

## Plotting monthly and yearly trends

```python
# get months from index
index_month=jobs.index.month
# rate by month - mean
jobs_by_month=jobs.groupby(index_month).mean()
#plot mean(s)
ax=jobs_by_month.plot(fontsize=6,linewidth=1)
# set x lbl
ax.set_xlabel('Month',fontsize=10)
# y lbl
ax.set_ylabel('Mean unemployment rate',fontsize=10)
# legend
ax.legend(bbox_to_anchor=(1,1.05),fontsize=10)
# plt
plt.show()
```

```python
# get years from index
index_year=jobs.index.year
# rate by years - mean
jobs_by_year=jobs.groupby(index_year).mean()
#plot mean(s)
ax=jobs_by_year.plot(fontsize=6,linewidth=1)
# set x lbl
ax.set_xlabel('Year',fontsize=10)
# y lbl
ax.set_ylabel('Mean unemployment rate',fontsize=10)
# legend
ax.legend(bbox_to_anchor=(1,1.05),fontsize=10)
# plt
plt.show()
```

## Applying time series decomposition

```python
# empty dictionary to store decomposition
jobs_decomp = {}
# column names
jobs_columns=jobs.columns
# iterate through all columns and apply decomposition and figure out things
for ts in jobs_columns:
    ts_decomposition = sm.tsa.seasonal_decompose(jobs[ts])
    jobs_decomp[ts]=ts_decomposition
```

```python
# check dictionary
print(jobs_decomp)
```

## Visualizing seasonality in multiple time series

```python
# use a different dicitonary for seasonal data
jobs_seasonal={}
# iterate over jobs_decomp and store seasonal data in jobs_seasonal
for ts in jobs_columns:
    jobs_seasonal[ts]=jobs_decomp[ts].seasonal
```

```python
# check jobs_seasonal
jobs_seasonal
```

```python
# create a pandas dataframe of seasonal data from dictionary
seasonality_df=pd.DataFrame.from_dict(jobs_seasonal)
#check df
seasonality_df.head()
```

```python
# remove label for index
seasonality_df.index.name=None
# create subplots of each classification
# plots
seasonality_df.plot(subplots=True, layout=(8,2),sharey=False,fontsize=9,linewidth=0.3,legend=False,figsize=(16,9))
#plot
plt.show()
```

## Correlation b/w multiple time series

```python
# compute spearman corr
seasonality_corr=seasonality_df.corr(method="spearman")
# clustermap of correlation
fig=sns.clustermap(seasonality_corr,annot=True,annot_kws={'size':9},linewidths=4,figsize=(15,10))
# set config of y labels
plt.setp(fig.ax_heatmap.yaxis.get_majorticklabels(),rotation=0)
# set config of x labels
plt.setp(fig.ax_heatmap.xaxis.get_majorticklabels(),rotation=90)
# show plot
plt.show()
```
